```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os


def process_csv(input_filename, output_folder):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(input_filename)

    # Initialize an empty list to store the processed data
    lst_xy = []

    # Process each row in the DataFrame
    for index, rows in df.iterrows():
        frame_id = rows['frame_id']
        action = rows['action']
        unclipped_reward = rows['unclipped_reward']

        # Iterate through columns starting from the 7th column
        for i in df.columns[6:]:
            col_index = df.columns.get_loc(i)

            # Skip columns before the 7th column
            if col_index % 2 == 0:
                y_pos = rows[i]
            elif col_index % 2 == 1:
                x_pos = rows[i]

                # Continue if either x_pos or y_pos is nan
                if np.isnan(x_pos) or np.isnan(y_pos):
                    continue

                lst_xy.append((x_pos, y_pos, frame_id, action, unclipped_reward))

    # Convert lst_xy into a DataFrame
    df_xy = pd.DataFrame(lst_xy, columns=['x_pos', 'y_pos', 'frame_id', 'action', 'score'])

    # Save df_xy into a new CSV file in the output folder
    output_filename = os.path.join(output_folder, f'{os.path.splitext(os.path.basename(input_filename))[0]}_xy.csv')
    df_xy.to_csv(output_filename, index=False)

# Specify the folder path
folder_path = 'ms_pacman/csv_files'

# Specify the output folder path
output_folder = 'cleaned_data'

# List all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Process each CSV file
for csv_file in csv_files:
    # Create the full path to the CSV file
    full_path = os.path.join(folder_path, csv_file)
    
    # Process the CSV file using the previously defined function
    process_csv(full_path, output_folder)



```
